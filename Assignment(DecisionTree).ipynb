```python
def outlier_capping(df,column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df[column]=df[column].apply(lambda x:lower_bound if x<lower_bound else upper_bound if x>upper_bound else x)
for col in df.select_dtypes(['int','float']).columns:
    outlier_capping(df,col)
```

# 3. Feature Engineering:

```python
cat_col=[i for i in feature.columns if feature[i].dtype=='O']
cat_col
```

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
```

```python
ord_enc= OrdinalEncoder()
```

```python
cat_col=pd.DataFrame(ord_enc.fit_transform(feature[['cp', 'restecg', 'exang', 'slope', 'thal']]),columns=feature[['cp', 'restecg', 'exang', 'slope', 'thal']].columns)
```

```python
cat_col
```

```python
df1=pd.get_dummies(feature[['sex']],columns=feature[['sex']].columns)
```

```python
df1
```

```python
df2=pd.DataFrame(ord_enc.fit_transform(df1),columns=df1.columns)
```

# 4. Decision Tree Classification:

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature_new,target,train_size=0.80,random_state=110)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dec_tree=DecisionTreeClassifier( criterion='gini', max_depth= 3, splitter= 'random')
```

```python
dec_tree.fit(x_train,y_train)
```

```python
y_pred=dec_tree.predict(x_test)
```

```python
y_pred
```

```python
print(f"Length of y_test: {len(y_test)}")
print(f"Length of y_pred: {len(y_pred)}")
```

```python
accuracy_score(y_test,y_pred)
```
