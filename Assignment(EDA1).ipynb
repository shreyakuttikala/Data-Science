```python
def outliers_capping(df,column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df[column]=df[column].apply(lambda x:lower_bound if x<lower_bound else upper_bound if x>upper_bound else x)
for col in df.select_dtypes(['int','float']).columns:
    outliers_capping(df,col)
    
```

# 3.	Data Visualization:

```python
num_col
```

```python
for col in num_col:
    sns.histplot(df[col])
    plt.title(f'Distribution of {col} (Histogram)')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
for col in num_col:
    sns.boxplot(y=df[col])
    plt.title(f'Distribution of {col} (Boxplot)')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
if df[cat_col].empty:
    print('No categorical columns')
else:
    for col in cat_col: 
        df[col].value_counts().plot(kind='bar')
        plt.title(f'Frequency of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df[num_col] ,x='LB', y='AC')
plt.title('Scatter Plot: Baseline FHR (LB) vs. Accelerations (AC)')
plt.xlabel('Baseline Fetal Heart Rate (LB)')
plt.ylabel('Accelerations (AC)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df[num_col] ,x='LB', y='ASTV')
plt.title('Scatter Plot: Baseline FHR (LB) vs. Abnormal Short Term Variability (ASTV)')
plt.xlabel('Baseline Fetal Heart Rate (LB)')
plt.ylabel('Percentage of Time with Abnormal Short Term Variability (ASTV)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df[num_col] ,x='LB', y='Width')
plt.title('Scatter Plot: Baseline FHR (LB) vs. Width')
plt.xlabel('Baseline Fetal Heart Rate (LB)')
plt.ylabel('Width')
plt.show()
```

```python

corr=df[num_col].corr()
```

```python
plt.figure(figsize=(13,9))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
selected_cols_for_pairplot = ['LB', 'AC', 'FM', 'UC', 'ASTV'] 
sns.pairplot(df[selected_cols_for_pairplot])
plt.suptitle('Pair Plot of Selected Features') 
plt.tight_layout()
plt.show()
```
