```python
numerical_features_after_preprocessing = df.select_dtypes(include=['int64', 'float64']).columns
skewed_features = df[numerical_features_after_preprocessing].skew().sort_values(ascending=False)
print("\nSkewness of numerical features:")
print(skewed_features)
```

```python
if 'capital.gain' in df.columns and df['capital.gain'].skew() > 1 and (df['capital.gain'] >= 0).all():
    df['capital.gain_log'] = np.log1p(df['capital.gain']) 
    print("\nApplied log transformation to 'capital.gain'. New feature 'capital.gain_log' created.")
elif 'fnlwgt' in df.columns and df['fnlwgt'].skew() > 1 and (df['fnlwgt'] >= 0).all():
    df['fnlwgt_log'] = np.log1p(df['fnlwgt'])
    print("\nApplied log transformation to 'fnlwgt'. New feature 'fnlwgt_log' created.")
else:
    print("\nNo highly skewed, positive numerical features found for log transformation, or 'capital.gain'/'fnlwgt' not present.")

print("\nFirst 5 rows of the dataset after log transformation (if applied):")
print(df.head())
```

```python
from sklearn.ensemble import IsolationForest
```

```python
numerical_features_for_outliers = df.select_dtypes(include=['int64', 'float64']).columns
print(f"\nFeatures used for Isolation Forest: {list(numerical_features_for_outliers)}")
```

```python
if not numerical_features_for_outliers.empty:
    iso_forest = IsolationForest(random_state=42)
    outlier_predictions = iso_forest.fit_predict(df[numerical_features_for_outliers])
    df['outlier'] = outlier_predictions

    # Remove outliers
    df_no_outliers = df[df['outlier'] == 1].drop(columns=['outlier'])
    print(f"\nOriginal dataset shape: {df.shape}")
    print(f"Dataset shape after outlier removal: {df_no_outliers.shape}")
    print(f"Number of outliers removed: {df.shape[0] - df_no_outliers.shape[0]}")
else:
    print("\nNo numerical features available for Isolation Forest outlier detection.")
    df_no_outliers = df.copy()
```
